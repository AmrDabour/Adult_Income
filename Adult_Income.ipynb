# Dealing with Categorical data

```python

categorical_data
```

```python

categorical_data.replace('?',np.nan,inplace=True)
```

```python
missing_vals = categorical_data.isna().sum()
missing_vals
```

```python
# categorical_data = categorical_data[~all_null]
# categorical_data
```

```python
# Replace '?' with the mode in each column
occupation_mode = categorical_data['occupation'].mode()[0]
workclass_mode = categorical_data['workclass'].mode()[0]
native_country_mode = categorical_data['native-country'].mode()[0]
categorical_data['occupation'] = categorical_data['occupation'].fillna( occupation_mode)
categorical_data['workclass'] = categorical_data['workclass'].fillna( workclass_mode)
categorical_data['native-country'] = categorical_data['native-country'].fillna( native_country_mode)

# Verify the replacement
categorical_data[['occupation', 'workclass', 'native-country']].head()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
categorical_data['occupation'] = lbl_encoder.fit_transform(categorical_data['occupation'])
categorical_data['workclass'] = lbl_encoder.fit_transform(categorical_data['workclass'])
categorical_data['native-country'] = lbl_encoder.fit_transform(categorical_data['native-country'])
categorical_data["education"] = lbl_encoder.fit_transform(categorical_data["education"])
categorical_data["marital-status"] = lbl_encoder.fit_transform(categorical_data["marital-status"])
categorical_data["relationship"] = lbl_encoder.fit_transform(categorical_data["relationship"])
categorical_data["race"] = lbl_encoder.fit_transform(categorical_data["race"])
categorical_data["gender"] = lbl_encoder.fit_transform(categorical_data["gender"])


```

```python
df = pd.concat([categorical_data,numerical_data],axis=1)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
corr_matrix = df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Draw the heatmap with the correlation matrix
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)

# Display the heatmap
plt.title('Correlation Heatmap')
plt.show()
```
